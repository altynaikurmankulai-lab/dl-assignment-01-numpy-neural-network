```python
#A.3 Activation Function (Sigmoid in Sales Forecasting)
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
test_values = np.array([-1, -3, 0, 3, 5])
print("Sigmoid output:")
print(sigmoid(test_values))
```

```python
#A.4 Gradient Calculation (Learning Adjustment)
def sigmoid_gradient(x):
    s = sigmoid(x)
    return s * (1 - s)
print("Sigmoid gradient:")
print(sigmoid_gradient(test_values))
```

```python
# C
import numpy as np
import matplotlib.pyplot as plt
# Dataset
X = np.array([
    [20, 15, 30, 25, 35],
    [3, 5, 2, 4, 2],
    [4, 3, 2, 1, 3]
])
Y = np.array([[18, 20, 22, 25, 30]])
# Activation functions
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)
def relu(z):
    return np.maximum(0, z)
def relu_derivative(z):
    return (z > 0).astype(float)
# Initialize parameters
np.random.seed(1)
W1 = np.random.randn(3,3) * 0.01
b1 = np.zeros((3,1))
W2 = np.random.randn(1,3) * 0.01
b2 = np.zeros((1,1))

learning_rate = 0.001
epochs = 5000
losses = []

for i in range(epochs):
    # Forward
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)

    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    # Loss
    loss = np.mean((A2 - Y)**2)
    losses.append(loss)
    # Backward
    dA2 = 2 * (A2 - Y)
    dZ2 = dA2 * sigmoid_derivative(Z2)

    dW2 = np.dot(dZ2, A1.T) / X.shape[1]
    db2 = np.sum(dZ2, axis=1, keepdims=True) / X.shape[1]

    dA1 = np.dot(W2.T, dZ2)
    dZ1 = dA1 * relu_derivative(Z1)

    dW1 = np.dot(dZ1, X.T) / X.shape[1]
    db1 = np.sum(dZ1, axis=1, keepdims=True) / X.shape[1]

    # Update
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

# Final predictions
Z1 = np.dot(W1, X) + b1
A1 = relu(Z1)
Z2 = np.dot(W2, A1) + b2
A2 = sigmoid(Z2)

print("Final W1:\n", W1)
print("\nFinal b1:\n", b1)
print("\nFinal W2:\n", W2)
print("\nFinal b2:\n", b2)

print("\nPredictions:")
print(A2)

# Plot loss
plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
